Spliting the train set 

```python
x=train.drop(columns=['id','label'],axis=1)
y=train['label']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = .25)
```

```python
m = LogisticRegression()
m.fit(X_train, y_train)
```

```python
m.score(X_test, y_test)
```

```python
from sklearn.neural_network import MLPClassifier
model= MLPClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model_1= GradientBoostingClassifier()
model_1.fit(X_train,y_train)
clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0).fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.svm import SVC

model_2 = SVC(probability=True)

model_2.fit(X_train,y_train)

model_2.score(X_test, y_test)
```

```python
model_2.get_params()
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(
    model_2,
    param_grid={
        'C': [5,7],
        'kernel': ['rbf', 'linear', 'poly'],
        'degree': [3,5,7],
        'gamma': ['scale', 'auto'],
    }, scoring='accuracy', cv=5, verbose=2
)
grid.fit(X_test, y_test)
```

```python
grid.best_params_
```

```python
from sklearn.model_selection import cross_val_score
cv = cross_val_score(model_2, X_test, y_test, cv=10)
cv
cv.mean()
```

```python
from keras.models import Sequential
from keras.layers import Dense, Input, Activation, BatchNormalization, Dropout
```

```python
model = Sequential()
model.add(Dense(128, input_dim=26, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=128)
```

```python
test_1 = test.drop(columns=['id'],axis=1)
```
